# Products

```python
df.head()
```

```python
#Full Category
df['FULL_CATEGORY'] = df['CATEGORY'].str.cat(df['SUB_CATEGORY'], sep='|').str.lower()

#Nombre de la tienda
df['STORE'] = "Richart's"
```

```python
#Quitamos etiquetas HTML
df['DESCRIPTION'] = df.DESCRIPTION.apply(lambda x: re.sub('<[^<]+?>', '', x))
```

```python
#Extraemos PACKAGE
df['PACKAGE'] = df['DESCRIPTION'].str.extract(r'(\d{1,4} ?[a-zA-Z]{1,3}.?$)')
```

```python
#Revisamos que colunas tienen valores vacios
df.columns[df.isna().any()].tolist()
```

```python
#Limpiamos cada columna deacuerdo al tipo de dato
df['BUY_UNIT'].fillna("None", inplace=True)
```

```python
df['DESCRIPTION_STATUS'].fillna("N", inplace=True)
```

```python
df['ORGANIC_ITEM'].fillna("Miss", inplace=True)
```

```python
df['NAME'].fillna("Miss", inplace=True)
```

```python
df['CATEGORY'].fillna("MISS", inplace=True)
```

```python
df['SUB_CATEGORY'].fillna("MISS", inplace=True)
```

```python
df['SUB_SUB_CATEGORY'].fillna("MISS", inplace=True)
```

```python
df['BRAND'].fillna("MISS", inplace=True)
```

```python
df['FULL_CATEGORY'].fillna("none", inplace=True)
```

```python
df['PACKAGE'].fillna("none", inplace=True)
```
